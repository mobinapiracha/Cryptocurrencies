```python
file_path = "data/shopping_data.csv"
df_shopping = pd.read_csv(file_path, encoding="ISO_8859-1")
df_shopping.head()
```

```python
print(df_shopping.columns)
print(df_shopping.dtypes)
```

```python
# Missing data 
for column in df_shopping.columns:
    print(f"Column {column} has {df_shopping[column].isnull().sum()} null values")
```

```python
# Drop all null values 
df_shopping = df_shopping.dropna()
```

```python
for column in df_shopping.columns:
    print(f"Column {column} has {df_shopping[column].isnull().sum()} null values")
```

```python
print(f"Duplicate entries: {df_shopping.duplicated().sum()}")
```

```python
df_shopping = df_shopping.drop(["CustomerID"], axis=1)
```

```python
# Transform categorical variables to binary 
df_shopping["Card Member"].head()
```

```python
def changestring(member): 
    if member == "Yes":
        return 1 
    else: 
        return 0 

df_shopping["Card Member"] = df_shopping["Card Member"].apply(changestring)
df_shopping.head()
```

```python
#Scale annual income 
df_shopping["Annual Income"] = df_shopping["Annual Income"]/1000
df_shopping.head()
```

```python
df_shopping.columns = ["Card_Member", "Age", "Annual_Income", "Spending_Score(1-100)"]
df_shopping.head()
```

```python
file_path = "data/shopping_data_cleaned.csv"
df_shopping.to_csv(file_path, index=False)
```
